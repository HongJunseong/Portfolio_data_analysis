### 프로젝트 목표 및 내용:

현대의 사람들은 각자의 여가 시간에 다양한 활동들을 한다. 그 활동들에는 운동하기, 음악듣기, 공부하기, 여행가기, 영화보기 같은 것들이 있다. 사람들은 이러한 활동들을 통해 쌓여 있던 스트레스를 해소하며 자신의 삶의 질을 높인다. 이 프로젝트는 앞서 말한 것들 중 '영화보기'라는 문화 생활을 좀 더 즐길 수 있도록 도움이 될만한 정보들을 사용자가 알기 쉽게하도록 하는 것이 목표이다. 사용자에게 볼만한 영화 추천, 각 영화에 대한 기타 정보들이 이 프로젝트에 담길 것 이다. 우리나라 영화를 시작으로 가능하다면 다른 나라의 영화들에 대한 정보도 담고자 한다.

### 주제 선정 이유 및 프로젝트의 필요성:

이렇게 바쁘게 돌아가는 세상 속에서도 사람들은 자신만의 여가 시간을 가지고 살아간다. 그렇게 자신만의 시간을 가진다는 것은 매우 중요한 일이다. 일, 학업 등에서 받는 스트레스들을 여가 시간을 잘 활용하여 해소하게 되면 일의 능률을 높일 수 있다는 연구 결과도 있을 정도이니 말이다.<br>
여러가지 여가 활동이 있는데 그중에서 영화보기라는 활동을 주제로 선정한데는 이유가 있다. 세계적으로는 하루에도 수백만 명의 사람들이 영화를 보고 있으며, 우리나라의 경우에도 천만관객을 돌파한 영화가 속속 등장하고 있다. 이를 통해 영화의 영향력이 얼마나 큰지 알 수 있다. 또한, 영화라는 것은 단순히 오락적 가치와 상품적 가치만 있는 것이 아니라, 사회적 가치를 갖기도 하고, 관객에게 심리적으로 영향을 주기도 한다. 예를 들어, 최근에 각종 영화제 수상을 휩쓴 영화'기생충'을 보면 보이지 않는 현대사회 계급의 구분인 돈, 청년실업 문제, 학벌주의, 물질만능주의 등 한국 사회의 암울한 면들이 고스란히 드러나고 있다. 이처럼 현대 사회의 어두운 면을 비추고 풍자할 수 있는, 영화라는 문화가 내게 매력적으로 다가왔다.<br>
평소에 영화를 자주 보지 않는 편이기 때문에, 이 프로젝트를 진행하면서 나도 영화들에 대해 좀 더 알아가고 싶기도 하고, 다른 사람들도 좀 더 영화에 대한 정보를 쉽게 접할 수 있었으면 하는 마음에서 이 프로젝트를 시작하게 되었다. 그리고, 특히 이렇게 전염병이 널리 퍼진 상황속에서는 외출을 최대한 자제하기 위하여 영화관에 가지 않고 집에서 볼 영화를 찾고자 하는 사람들이 많을 것이라고 생각했기 때문에 이 주제를 선정하게 되었다.

영화 흥행작의 기준을 1000만 관객, 500만 관객이라고 생각하였다. 어떤 영화들의 관객 수가 1000만을 넘었고, 500만을 넘었는지 확인하고자 하였다.
아래의 코드는, '1000만'을 입력하면, 관객 수가 1000만 명이 넘었던 영화의 목록을 보여주고, '500만'을 입력하면, 관객 수가 500만 ~ 1000만 명이였던 영화의 목록을 보여준다.

```python
import pandas as pd

movie_aud = pd.read_excel("movie_audience.xlsx")
movie_aud = movie_aud.sort_values(by='관객수', ascending=False)

def movaud():
    aud = input("영화 관객수의 기준을 입력해 주세요(ex. 1000만, 500만) : ")
    if aud == '1000만':
        result = movie_aud[movie_aud['관객수'] >= 10000000]
    elif aud == '500만':
        result = movie_aud[movie_aud['관객수'] < 10000000]

    result = result.reset_index(drop=True)
    result.index += 1
    return result

display(movaud())
```

```python
import numpy as np
from konlpy.tag import Okt
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image

# 파일 읽기
with open("movie_cloud.txt", 'r', encoding='UTF-8') as file:
    lists = file.readlines()

# Okt 사용 (Twitter가 Okt로 이름이 변경됨)
okt = Okt()
allword = []

for txt in lists:
    allword.append(okt.pos(txt))

clearli = []
for clword in allword:
    for word, tag in clword:
        if tag == 'Noun' and word not in ["것", "내", "나", "수", "게", "말", "네이버", "때"]:
            clearli.append(word)

count = Counter(clearli)
words = dict(count.most_common())

alice_mask = np.array(Image.open("alice_mask.png"))

wordcloud = WordCloud(font_path='C://Windows//Fonts//gulim',
                      background_color='black',
                      max_words=200,
                      mask=alice_mask,
                      colormap="Accent_r").generate_from_frequencies(words)

plt.figure(figsize=(11, 9))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```

위의 결과는 '영화'라는 키워드를 검색을 했을 때 나타나는 결과를 크롤링하여, '영화'라는 단어와 가장 연관성이 있는 단어를 토대로 생성한 wordcloud 이다.<br>

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

def get_movie_summary(movname):
    # Selenium webdriver 설정
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')  # Headless 모드
    options.add_argument('--disable-gpu')
    driver = webdriver.Chrome(options=options)
    
    try:
        # Google에서 영화 검색
        driver.get('https://www.google.com/search?q=' + movname + "+영화 줄거리")
        time.sleep(2)  # 페이지 로드 대기
        
        # HTML 파싱
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        
        # 영화 줄거리 찾기
        summary = None
        
        # 첫 번째 시도: 유명한 영화일 경우
        try:
            sour = soup.find('div', attrs={'class': 'kno-rdesc'})
            if sour:
                summary = sour.get_text().strip("설명더보기")
        
        # 예외 처리
        except Exception as e:
            print(f"첫 번째 시도에서 오류 발생: {e}")
        
        # 두 번째 시도: 유명하지 않은 영화일 경우
        if not summary:
            try:
                soups = soup.find('div', attrs={'class': 'twQ0Be'})
                if soups:
                    element = soups.find("a")
                    link = element.get("href")
                    driver.get(link)
                    time.sleep(2)
                    
                    html = driver.page_source
                    soup = BeautifulSoup(html, 'html.parser')
                    
                    summary_div = soup.find('div', attrs={'class': 'some-class-for-summary'})  # 클래스명 변경 필요
                    if summary_div:
                        summary = summary_div.get_text().strip()
            
            # 예외 처리
            except Exception as e:
                print(f"두 번째 시도에서 오류 발생: {e}")
        
        # 최종 결과 출력
        if summary:
            print("\n영화 줄거리: ", summary)
        else:
            print("영화 줄거리를 찾을 수 없습니다.")
    
    finally:
        driver.quit()

#영화 제목 검색
movname = input("영화의 제목을 입력해주세요 : ")
get_movie_summary(movname)
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

def get_movie_trailer(movname):
    # Selenium webdriver 설정
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')  # Headless 모드
    options.add_argument('--disable-gpu')
    driver = webdriver.Chrome(options=options)
    
    try:
        # Google에서 영화 예고편 검색
        driver.get('https://www.google.com/search?q=' + movname + "+영화+예고편")
        time.sleep(2)  # 페이지 로드 대기
        
        # HTML 파싱
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        
        # 예고편 링크 찾기
        link = None
        
        # 첫 번째 시도: 유명한 영화일 경우
        try:
            soups = soup.find('div', attrs={'class': 'twQ0Be'})
            if soups:
                element = soups.find("a")
                link = element.get("href")
        
        # 예외 처리
        except Exception as e:
            print(f"첫 번째 시도에서 오류 발생: {e}")
        
        # 두 번째 시도: 유명하지 않은 영화일 경우
        if not link:
            try:
                soups = soup.find('a', attrs={'class': 'WpKAof'})
                if soups:
                    link = soups.get("href")
            
            # 예외 처리
            except Exception as e:
                print(f"두 번째 시도에서 오류 발생: {e}")
        
        # 최종 결과 출력
        if link:
            print("영화 예고편 링크: ", link)
        else:
            print("영화 예고편을 찾을 수 없습니다.")
    
    finally:
        driver.quit()

#영화 제목 검색
movname = input("영화의 제목을 입력해주세요 : ")
get_movie_trailer(movname)
```

위 코드는, 예고편이 궁금한 영화의 제목을 입력하면 해당 영화의 예고편이 담긴 유튜브 주소를, selenium을 통하여 크롤링하는 코드이다.

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import webbrowser
import time

def movact_1():
    # 사용자로부터 영화 제목 입력 받기
    movname = input("영화의 제목을 입력해주세요: ")
    print()

    # Selenium webdriver 설정
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')  # Headless 모드
    options.add_argument('--disable-gpu')
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    
    try:
        # Google에서 영화 출연진 검색
        driver.get('https://www.google.com/search?q=' + movname + "+영화+출연진")
        time.sleep(2)  # 페이지 로드 대기
        
        # HTML 파싱
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        
        # 출연진 목록 찾기
        soups = soup.find_all('div',attrs={'class':'JjtOHd'})
        
        # 출연진 출력
        if soups:
            print("출연진 목록:")
            for tag in soups:
                print(tag.text)
        else:
            print("출연진 정보를 찾을 수 없습니다.")
            return
        
        print()    
        # 사용자로부터 궁금한 인물 이름 입력 받기
        actname = input("위의 출연진들 중에서 궁금한 인물의 이름을 입력해주세요: ")
        print()
        
        # 웹 브라우저에서 해당 인물의 Google 검색 결과 열기
        webbrowser.open('https://www.google.com/search?q=' + actname)
    
    finally:
        driver.quit()

# 입력한 영화의 출연진 출력
movact_1()
```

위 코드는, 출연진이 궁금한 영화의 제목을 입력하면 해당 영화의 출연진들의 목록을 selenium을 통하여 크롤링하는 코드이다. 이렇게 출력된 출연진들 중에서 더 자세히 알고 싶은 배우의 이름을 입력하면, 새로운 웹페이지에서 해당 배우를 검색한 결과가 나타난다.

```python
import matplotlib.pyplot as plt

def display_genre_recommendations(genre):
       
    # 영화 장르 구글 검색 링크
    genre_links = {
        "액션": "https://www.google.com/search?q=액션+장르+영화",
        "로맨스": "https://www.google.com/search?q=로맨스+장르+영화",
        "애니": "https://www.google.com/search?q=애니+장르+영화",
        "음악": "https://www.google.com/search?q=음악+영화",
        "코미디": "https://www.google.com/search?q=코미디+장르+영화",
        "공포": "https://www.google.com/search?q=공포+장르+영화"
    }
    
    if genre in genre_links:
        
        print(f"이 장르의 영화를 더 알고 싶다면..? 링크: {genre_links[genre]}")
    else:
        print("지원하지 않는 장르입니다. 다음 중에서 선택해주세요: 액션, 로맨스, 애니, 음악, 코미디, 공포")

# 사용 예시
genre = input("어떤 장르의 영화를 원하시나요? : ")
display_genre_recommendations(genre)
```

----
사용자가 원하는 장르를 입력하면, 입력한 장르에 해당하는 영화 목록이 담긴 웹사이트 링크를 출력하여 바로 확인할 수 있도록 한다.
